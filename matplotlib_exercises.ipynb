```python


# Generate x values
x = np.arange(-11, 11)  # Generates x values from -10 to 10 (inclusive)

# Compute y values
y = x**2 - x + 2

# Plot the equation
plt.plot(x, y)

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2 - x + 2')

# Show the plot
plt.show()
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y=√x

You can use functions from the math module to help implement some of the equations above.

```python
# Generate x values
x1 = np.arange(-10, 10)  # Generates x values from -10 to 10 (inclusive)

# Compute y values
y1 = np.sqrt(x1)


# Plot the equation
plt.plot(x1, y1)

# Set labels and title
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('Plot of y=√x')

# Show the plot
plt.show()
```

```python
# Generate x values
x2 = np.arange(-100, 100)  # Generates x values from -10 to 10 (inclusive)

# Compute y values
y2 = [num**3 for num in x2]


# Plot the equation
plt.plot(x2, y2)

# Set labels and title
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('Plot of y=x^3')

# Show the plot
plt.show()
```

```python
x3 = np.arange(0, 200)  # Generates x values from -10 to 10 (inclusive)

# Compute y values
y3 = [2**num for num in x3]


# Plot the equation
plt.plot(x3, y2)

# Set labels and title
plt.xlabel('x3')
plt.ylabel('y3')
plt.title('Plot of y=2^x')

# Show the plot
plt.show()
```

```python
x4 = np.arange(0, 10) # Generates x values from -10 to 10 (inclusive)

# Compute y values
y4 = [(1/num + 1) for num in x4]


# Plot the equation
plt.plot(x4, y4)

# Set labels and title
plt.xlabel('x4')
plt.ylabel('y4')
plt.title('Plot of y=1/(x+1)')

# Show the plot
plt.show()
```
